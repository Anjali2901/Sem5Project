```python
import pandas as pd
df = pd.read_csv("outfile2.csv")
li = []
for index,row in df.iterrows():
    lat = "("+str(row['latitude'])
    long = str(row['longitude'])+")"
    location = lat+","+long
    li.append(location)
li
se = pd.Series(li)
df['Location'] = se.values
df
```

```python
from textblob import TextBlob
li = []
for i in df["text"]:
    analysis = TextBlob(i)
    if analysis.sentiment.polarity > 0:
        li.append(0)
    elif analysis.sentiment.polarity < 0:
        li.append(1)
    else:
        li.append(1)
se = pd.Series(li)
df['Sentiment'] = se.values
df = df.sort_values(['Sentiment'], ascending=1)
df.drop(labels = ['id','Time','user','Location'], inplace =True, axis = 1)
df.to_csv("senti_.csv")
```

```python
def tokenize(text): 
    tknzr = TweetTokenizer()
    return tknzr.tokenize(text)

def stem(doc):
    return (stemmer.stem(w) for w in analyzer(doc))

en_stopwords = set(stopwords.words("english")) 

vectorizer = CountVectorizer(
    analyzer = 'word',
    tokenizer = tokenize,
    lowercase = True,
    ngram_range=(1, 1),
    stop_words = en_stopwords)
```

```python
kfolds = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
```

```python
np.random.seed(1)

pipeline_svm = make_pipeline(vectorizer,SVC(probability=True, kernel="linear", class_weight="balanced"))

grid_svm = GridSearchCV(pipeline_svm,
                        param_grid = {'svc__C': [0.01, 0.1, 1]},
                        cv = kfolds,
                        scoring = "roc_auc",
                        verbose=1,
                        n_jobs=-1) 

grid_svm.fit(X_test, y_test)
grid_svm.score(X_test, y_test)
```
