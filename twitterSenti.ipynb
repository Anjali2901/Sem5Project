```python
import pandas as pd
df = pd.read_csv("outfile2.csv")
li = []
for index,row in df.iterrows():
    lat = "("+str(row['latitude'])
    long = str(row['longitude'])+")"
    location = lat+","+long
    li.append(location)
li
se = pd.Series(li)
df['Location'] = se.values
df
```

```python
from textblob import TextBlob
li = []
for i in df["text"]:
    analysis = TextBlob(i)
    if analysis.sentiment.polarity > 0:
        li.append(1)
    elif analysis.sentiment.polarity < 0:
        li.append(-1)
    else:
        li.append(0)
se = pd.Series(li)
df['Sentiment'] = se.values
df = df.sort_values(['Sentiment'], ascending=1)
```

```python
df1 = pd.DataFrame()

li = []
li1 = []
for index,row in df.iterrows():
    if row['Sentiment'] < 0.05:
        li.append(row['latitude'])
        li1.append(row['longitude'])
df1 = pd.DataFrame({'latitude':li,'longitude':li1})
df1
```

```python
df1.to_csv('people.csv', sep=',' , index=None)
```
